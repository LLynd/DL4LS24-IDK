## Merkers expression distribution in cell types

```python
for marker in markers:
    fig, ax = plt.subplots(5, 3, figsize=(15, 10))
    fig.suptitle(marker)

    for i, celltype in enumerate(celltypes):
        m_expr = df_markers[df_markers.cell_label == celltype][marker]
        ax[i//3, i%3].hist(m_expr, bins=50, alpha=0.8)
        ax[i//3, i%3].set_title(celltype)

    plt.savefig(f'graphics/{marker}_hists.png')
    plt.close(fig)
    
```

```python
chosen_markers = ['CD15', 'CD163', 'CD206', 'Ecad', 'FOXP3', 'MPO']
for marker in chosen_markers:
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.suptitle(f'{marker} expression distribution across cell types')

    for celltype in celltypes:
        m_expr = df_markers[df_markers.cell_label == celltype][marker]
        ax.hist(m_expr, bins=50, alpha=0.4, density=True, label=celltype)

    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.set_facecolor('#f7f7f7')
    ax.set_xlabel('Expression')
    ax.legend()
    plt.savefig(f'graphics/{marker}_density_hists.png')
    plt.close(fig)
```

## Loading images

```python
# adata - anndata object

image_name = adata.obs.iloc[20000]['image']
image_path = os.path.join('train/images_masks/img', image_name)
mask_path = os.path.join('train/images_masks/masks', image_name)

image_reader = pyometiff.OMETIFFReader(fpath=image_path)
mask_reader = pyometiff.OMETIFFReader(fpath=mask_path)

image_array, _, _ = image_reader.read()
mask_array, _, _ = mask_reader.read()

print(image_array.shape)
plt.imshow(mask_array)

for i in range(image_array.shape[0]):
    plt.imshow(np.arcsinh(image_array[i] / 5.))
    plt.savefig(f'graphics/imc_img/{markers[i]}.png')
    plt.close()
```
